```python
import pandas as pd
import numpy as np
import torch
import os
from datasets import load_dataset
from transformers import pipeline
from dotenv import load_dotenv
```

```python
load_dotenv()
device = "cuda" if torch.cuda.is_available() else "cpu"

pipe = pipeline("text-generation", model="openai-community/gpt2", token=os.getenv("HF_TOKEN"), device=device)
```

```python
responses = pipe('Do you know who Anthony Hopkins is?', max_length=2, num_return_sequences=3)

for response in responses:
    print(response['generated_text'])
```
