```python
import pandas as pd
import numpy as np
import torch
import os
from transformers import pipeline
from dotenv import load_dotenv
```

```python
load_dotenv()
pipe = pipeline("text-generation", model="openai-community/gpt2", token=os.getenv("HF_TOKEN"))
```

```python
responses = pipe('Do you know who Anthony Hopkins is?', max_length=2, num_return_sequences=2)

for response in responses:
    print(response['generated_text'])
```
