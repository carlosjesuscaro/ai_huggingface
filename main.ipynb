```python
import pandas as pd
import numpy as np
import torch
import os
import tqdm as notebook_tqdm
from transformers import pipeline
from dotenv import load_dotenv
```

```python
load_dotenv()
pipe = pipeline("text-generation", model="openai-community/gpt2", token=os.getenv("HF_TOKEN"))
```

```python
responses = pipe('What is the capital of France?', max_length=50, num_return_sequences=30)

for response in responses:
    print(response['generated_text'])
```
